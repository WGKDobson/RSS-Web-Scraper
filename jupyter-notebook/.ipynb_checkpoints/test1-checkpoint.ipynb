```python
import requests
from bs4 import BeautifulSoup
import pickle

#url = "http://feeds.bbci.co.uk/news/world/rss.xml"
# url = "https://www.nytimes.com/section/world/rss.xml"
#url = "http://rssfeeds.usatoday.com/usatoday-newstopstories&x=1"
# url = "http://www.washingtonpost.com/rss/"
# --- xml urls ---
#url = "http://rss.cnn.com/rss/cnn_topstories.rss"
#url = "http://rss.cnn.com/rss/cnn_topstories.rss"
#url = "http://feeds.nytimes.com/nyt/rss/HomePage"
#url = "http://www.npr.org/rss/rss.php?id=1001"
#url = "http://newsrss.bbc.co.uk/rss/newsonline_world_edition/americas/rss.xml"
rssurls = {'bbc':'http://newsrss.bbc.co.uk/rss/newsonline_world_edition/americas/rss.xml', 'nyt':'http://feeds.nytimes.com/nyt/rss/HomePage', 'npr':'http://www.npr.org/rss/rss.php?id=1001'}

```

```python
for key, url in rssurls.items():
    resp = requests.get(url)
    soup = BeautifulSoup(resp.content, 'xml')
    items = soup.findAll('item')
    
    #print(soup.prettify())
    news_items = []   # list of news items
    for item in items:
        news_item = {}
        news_item['title'] = item.title.text
        news_item['description'] = item.description.text
        news_item['link'] = item.link.text
        news_items.append(news_item)
    print(str(key)+" news items read = "+str(len(items)))
    fname = "news_"+str(key)+"_today.p"
    pickle.dump(news_items, open(fname, "wb"))
    print("File:   " + fname + " ... Pickled")
   
    
```
